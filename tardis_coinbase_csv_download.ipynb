```python
count = 0
threshold = pytz.UTC.localize(datetime.datetime(2020, 5, 1))
token_wanted = "BTC"
symbols_list = []
for item in coinbase_details['availableSymbols']:
    raw_time = item['availableSince']
    parsed_time = dateutil.parser.isoparse(raw_time)
    pair_name = item['id']
    if parsed_time < threshold and token_wanted in pair_name:
#         print(pair_name)
        symbols_list.append(pair_name)
        count += 1
print(count)
print(symbols_list)
```
